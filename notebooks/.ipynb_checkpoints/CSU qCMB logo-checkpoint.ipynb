```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# mean, std dev
s = pd.DataFrame({'normal1': np.random.normal(0, 20, 1500),
                  'normal2': np.random.normal(90, 30, 1500),
                  'normal3': np.random.normal(180, 40, 1500),
                 })

melt = s.melt(var_name='distribution type', value_name='distributions')
```

```python

ax = sns.distplot(s['normal1'], hist=False, kde_kws={"shade": True, 'color':'#215732', 'alpha':0.5},)
#                   hist_kws={'edgecolor':'black', 'linewidth':2})

ax = sns.distplot(s['normal2'], hist=False, kde_kws={"shade": True, 'color':'#d9d499', 'alpha':0.5},)
#                   hist_kws={'edgecolor':'black', 'linewidth':1})

ax = sns.distplot(s['normal3'], hist=False, kde_kws={"shade": True, 'color':'white', 'alpha':0.5},)
#                   hist_kws={'edgecolor':'black', 'linewidth':1})

plt.ylabel('')
plt.xlabel('')

plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, labelleft=False,)
ax.set_xticklabels(labels=[''])
plt.xlabel('')
plt.ylabel('')
sns.despine(left=True)

plt.savefig('CSU qCMB logo test.png', dpi=1000)
```
