```python
# generating ormal distributions specified by mean, std deviation, # observations

s = pd.DataFrame({'normal1': np.random.normal(0, 2, 100),
                  'normal2': np.random.normal(5, 4, 100),
                  'normal3': np.random.normal(10, 6, 100),
                 })

# melting in case..
melt = s.melt(var_name='distribution type', value_name='distributions')
```

```python
#moving the rain under the boxplot
f, ax = plt.subplots(figsize=(12, 5))

x = 'distribution type'
y = 'distributions'
hue = None
data = melt
sigma = 0.1
pal = {'normal1': '#215732', 'normal2': '#d9d499', 'normal3':'grey'}

ax=ptp.RainCloud(x = x, y = y, hue = hue, data = data, palette = pal, bw = sigma, width_viol = .7,
                ax = ax, orient = 'h' , alpha = 1, dodge = True, pointplot = False, move = .2)

plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, labelleft=False,)
ax.set_xticklabels(labels=[''])
plt.xlabel('')
plt.ylabel('')
sns.despine(left=True)

plt.savefig('CSU qCMB logo test_2.png', dpi=1000)
```
